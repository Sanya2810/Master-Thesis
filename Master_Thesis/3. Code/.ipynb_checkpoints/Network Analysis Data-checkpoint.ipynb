```python
import pandas as pd
import os
```

```python
directory = os.getcwd()
url = directory[:-7] + '2. Data/'
```

```python
portfolio = pd.read_excel(url + 'ECB Portfolio - Company Information copy.xlsx')
portfolio.head()
```

```python
columns_to_keep = ['ISIN_CODE','ISSUER_NAME','COUPON_RATE','MATURITY_DATE','PUBLISHED_DATE']
isin_code = portfolio[columns_to_keep]
isin_code = isin_code.drop_duplicates(subset = columns_to_keep)
```

```python
columns = ['ISSUER_NAME','NACE_CODE','SECTOR','NCB']
company = portfolio[columns]
company = company.drop_duplicates(subset = columns)
company.head()
```

```python
columns = ['NACE_CODE','SECTOR']
sector = portfolio[columns]
sector = sector.drop_duplicates(subset = columns)
sector
```

```python
country = portfolio[['NCB']]
country = country.drop_duplicates(subset = ['NCB'])
country
```

```python
isin_code.to_csv(url + 'isin_code_node.csv',index=False)
```

```python
company.to_csv(url + 'company_node.csv',index=False)
sector.to_csv(url + 'sector_node.csv',index=False)
country.to_csv(url + 'country_node.csv',index=False)
```
