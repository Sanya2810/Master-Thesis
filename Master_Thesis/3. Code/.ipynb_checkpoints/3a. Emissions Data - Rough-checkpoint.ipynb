```python
import pandas as pd
import os
```

```python
directory = os.getcwd()
url = directory[:-7] + "2. Data/3. Emissions Data - CDP/GHG_Emissions_Data - CDP/"
```

```python
company_info = pd.read_excel(url + "Investor CDP 2017_Public Data (with sector Qs).xlsx", sheet_name = "Summary Data")
```

```python

emissions_baseyr = pd.read_excel(url + "Investor CDP 2017_Public Data (with sector Qs).xlsx", sheet_name = "CC7.1",skiprows=1,header=None)


```

```python
company_info = company_info[['account_id','account_name','isin']]
company_info.head()
```

```python
emissions_baseyr = emissions_baseyr.iloc[:,[1,6,9,10,11]]
emissions_baseyr.columns = ['ACCOUNT_ID','BASE_YEAR','SCOPE1_BASEYR','SCOPE2_BASEYR_LOCATION','SCOPE2_BASEYR_MARKET']
emissions_baseyr.head()
```

```python
emissions_global_scope1 = pd.read_excel(url + "Investor CDP 2017_Public Data (with sector Qs).xlsx", sheet_name = "CC8. Emissions Data",skiprows = 1,header = None)
```

```python
emissions_global_scope1.head()
```

```python
emissions_global_scope1= emissions_global_scope1.iloc[:,[1,5,7]]
emissions_global_scope1.columns = ['ACCOUNT_ID', 'ACCOUNTING_YEAR','SCOPE1_GLOBAL']
emissions_global_scope1.head()
```

```python
df = emissions_global_scope1.set_index(['ACCOUNT_ID','ACCOUNTING_YEAR']).stack().unstack(1)
df = df.rename_axis(('Account_ID','Level')).reset_index()
df = df.drop('Level',axis = 1)
df.columns = [str(col) + '_scope1_global' for col in df.columns]
df = df.rename(columns = {'Account_ID_scope1_global':'Account_Id'})
df.head()
```

```python
emissions_global_scope2 = pd.read_excel(url + "Investor CDP 2017_Public Data (with sector Qs).xlsx", sheet_name = "CC8.3a",skiprows = 2, header = None)
```

```python
emissions_global_scope2= emissions_global_scope2.iloc[:,[1,5,6,7]]
emissions_global_scope2.columns = ['ACCOUNT_ID', 'Accounting_Year','Location_Based','Market_Based']
emissions_global_scope2.head()
```

```python
df2 = emissions_global_scope2.set_index(['ACCOUNT_ID','Accounting_Year']).stack().unstack(1)
df2 = df2.rename_axis(('Account_Id','Type')).reset_index()
df2.columns = [str(col) + '_scope2_global' for col in df2.columns]
df2 = df2.rename(columns = {'Account_Id_scope2_global':'Account_Id'})
df2.head()
```

```python
df = emissions_global_scope2.set_index(['ACCOUNT_ID','ACCOUNTING_YEAR']).stack().unstack(1)
df = df.rename_axis(('Account_ID','Level')).reset_index()
df = df.drop('Level',axis = 1)
df.columns = [str(col) + '_scope1_global' for col in df.columns]
df = df.rename(columns = {'Account_ID_scope1_global':'Account_Id'})
df.head()
```
